# With the correct normalizations, signal region can be optimized : two correlated discriminating variables Mjj,dYjj, 2D square cuts are shown below.

```python
keys=['data','wenugvbs','wenug','wenu','zee','dp','ttbar','tg','dp_mc','wenu_mc']
Mjj_cuts=[200,250,300,350,400,450,500,550,600,650,700,750,800,850,900,950,1000]
dYjj_cuts=[1.,1.5,2.0,2.5]
opt_list = []
plot_title = 'No PtBalanceCut,Pt_LeadingJet>30,Pt_SubleadingJet>30'
plot_tag = 'Bveto70_NoPtBalance_jet0_30_jet1_30'
output = TwoJet_sig(csvfile='/Users/Mia/Documents/Work/WgZgPhysics/Analysis/WgTemplates/0807_2014_VeryTight/wenu_mjj_filter.csv')
excluding_wenug_sample = 'wenug'
wenu_vbsfilter_norm = 4999285/1.715635940664000e+16

for Mjj_cut in Mjj_cuts:
    for dYjj_cut in dYjj_cuts:
        # apply vbs cuts to each sample
        vbs_sig_region = {key:apply_vbs_cuts(sample=item,LeadingJetPt=30,SubleadingJetPt=30,
                                             MjjCut=Mjj_cut,dYjjCut=dYjj_cut,PtBalanceCut=.1) 
                                             for key,item in outputs.iteritems()}
        # calculate the sum of weights in vbs region
        vbs_sig_region_components = {key:ufloat(sum(item['weight']),sqrt(sum((item['weight']**2)))) 
                                     for key, item in vbs_sig_region.iteritems()}
        # normalize to the right normalization (for DD methods)
        vbs_sig_region_components_final = {key:item/normalization[key] for key, item in vbs_sig_region_components.iteritems()}
        vbs_sig_region['wenu_vbsfilter']=apply_vbs_cuts(sample=output,MjjCut=Mjj_cut,dYjjCut=dYjj_cut,PtBalanceCut=.1)
        vbs_sig_region_components['wenu_vbsfilter'] = ufloat(sum(vbs_sig_region['wenu_vbsfilter']['mc_event_weight']
                                                      *vbs_sig_region['wenu_vbsfilter']['weight']*wenu_vbsfilter_norm),\
                                                      sqrt(sum((vbs_sig_region['wenu_vbsfilter']['mc_event_weight']
                                                      *vbs_sig_region['wenu_vbsfilter']['weight']*wenu_vbsfilter_norm)**2)))
   
        vbs_sig_region_components_final['wenu_vbsfilter'] = vbs_sig_region_components['wenu_vbsfilter']

        # calculate total bkg and signal
        signal = vbs_sig_region_components_final['wenugvbs']
        background = sum([0 if el[0] in ['data','wenu_mc','dp_mc','wenugvbs','wenu_vbsfilter']\
                          else el[1] for el in vbs_sig_region_components_final.items()])\
                          #+vbs_sig_region_components_final['wenug_vbsfilter']
        
        # recording information of cuts used, signal,background,significance
        vbs_sig_region_components_final['Mjj'] = Mjj_cut
        vbs_sig_region_components_final['dYjj'] = dYjj_cut
        vbs_sig_region_components_final['background'] = background
        vbs_sig_region_components_final['signal'] = signal
        vbs_sig_region_components_final['signal_obv'] = vbs_sig_region_components_final['data']-background
        vbs_sig_region_components_final['sig_ratio'] = signal/background
        vbs_sig_region_components_final['significance'] = signal/sqrt(background)
        vbs_sig_region_components_final['significance_obv'] = signal/sqrt(vbs_sig_region_components_final['data']-signal)
        vbs_sig_region_components_final['data_mc_ratio'] = vbs_sig_region_components_final['data']/(signal+background)
        opt_list.append(vbs_sig_region_components_final)

x= [item['Mjj'] for item in opt_list if item['dYjj'] == 1.]   

errorplot([item['Mjj'] for item in opt_list if item['dYjj'] == 1.],
          [item['sig_ratio'] for item in opt_list if item['dYjj'] == 1.],label='dYjj > 1.0')
errorplot([item['Mjj'] for item in opt_list if item['dYjj'] == 1.],
          [item['sig_ratio'] for item in opt_list if item['dYjj'] == 1.5],label='dYjj > 1.5')
errorplot([item['Mjj'] for item in opt_list if item['dYjj'] == 1.],
          [item['sig_ratio'] for item in opt_list if item['dYjj'] == 2.],label='dYjj > 2.0')
errorplot([item['Mjj'] for item in opt_list if item['dYjj'] == 1.],
          [item['sig_ratio'] for item in opt_list if item['dYjj'] == 2.5],label='dYjj > 2.5')
legend(loc=0,bbox_to_anchor = (0.3, 0.85),fontsize=8)

title(plot_title)
xlabel('Mjj [GeV]')
ylabel('S/B')
savefig('/Users/Mia/Documents/Work/Slides/WgVBS/SigRatio_'+plot_tag+'.pdf')
clf()
errorplot([item['Mjj'] for item in opt_list if item['dYjj'] == 1.],
          [item['significance'] for item in opt_list if item['dYjj'] == 1.],label='dYjj > 1.0')
errorplot([item['Mjj'] for item in opt_list if item['dYjj'] == 1.],
          [item['significance'] for item in opt_list if item['dYjj'] == 1.5],label='dYjj > 1.5')
errorplot([item['Mjj'] for item in opt_list if item['dYjj'] == 1.],
          [item['significance'] for item in opt_list if item['dYjj'] == 2.]  ,label='dYjj > 2.0')
errorplot([item['Mjj'] for item in opt_list if item['dYjj'] == 1.],
          [item['significance'] for item in opt_list if item['dYjj'] == 2.5]  ,label='dYjj > 2.5')
legend(loc=0,bbox_to_anchor = (0.6, 0.3),fontsize=8)
title(plot_title)
xlabel('Mjj [GeV]')
ylabel('S/$\sqrt{B}$')
savefig('/Users/Mia/Documents/Work/Slides/WgVBS/Significance_'+plot_tag+'.pdf')  
```
