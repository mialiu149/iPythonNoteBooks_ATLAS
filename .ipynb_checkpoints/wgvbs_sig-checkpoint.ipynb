```python
%pylab inline
import numpy as np
import rpy2
```

```python
data = np.genfromtxt('/Users/Mia/iPythonNoteBooks/data/wgvbs/significance_pt20_jet050.csv', delimiter=',', skip_header=10,
                     skip_footer=10, names=['Mjj','dYjj','wgdata','wg','wgvbs','wjets','z','dp','ttbar'])
sig_ratio=data['wgvbs']/(data['wg']+data['wjets']+data['z']+data['dp']+data['ttbar'])
significance=data['wgvbs']/sqrt((data['wg']+data['wjets']+data['z']+data['dp']+data['ttbar']))
#sig_ratio=data['wgvbs']/(data['wgdata']-data['wgvbs'])
#significance=data['wgvbs']/sqrt(data['wgdata']-data['wgvbs'])
```

```python
for increment in range(29)[1:28:5]:
    print increment
    errorbar(data[increment:increment+4]['dYjj'],significance[increment:increment+4],fmt='--o',label=str(data[increment]['Mjj'])+' GeV')
    xlim(0.5,3.5)
   # ylabel('r$\frac{S}{\sqrt{B}}$')
    ylabel('S/(B)^(1/2)')
    xlabel('dYjj cut')
annotate('local max', xy=(2, 1), xytext=(3, 1.5),arrowprops=dict(facecolor='black', shrink=0.05),)
legend(loc=0,bbox_to_anchor = (0.75, 0.7),fontsize=8,title='Mjj Cut')
title('With Pt(jet) > 20GeV, leading jet pt > 50 GeV')
savefig('/Users/Mia/iPythonNoteBooks/plots/wgvbs/sigma_pt20_def1_jet0pt50.pdf')
```
