```python
%pylab inline
from datetime import datetime, date, time
import numpy as np
fcooling=open("/Users/Mia/Documents/SCTupgrade/CO2_Cooling/20120330.txt")
outputlocation = '/Users/Mia/Documents/SCTupgrade/CO2_Cooling/'
tinside = []
hinside = []
houtside = [] 
toutside = []
times = []
dewpointinside = []
dewpointoutside = []
#start = '11:29:51'
start = '11:31:57'
starttime=datetime.strptime(start,"%H:%M:%S")
#from sympy import *
```

```python
 
 # approximation valid for
# 0 degC < T < 60 degC
# 1% < RH < 100%
# 0 degC < Td < 50 degC 
 
# constants
a = 17.271
b = 237.7 # degC
 
# sys.argv[0] is program name 
def dewpoint_approximation(T,RH): 
    Td = (b * gamma(T,RH)) / (a - gamma(T,RH))
    return Td
  
def gamma(T,RH): 
    g = (a * T / (b + T)) + np.log(RH/100.0)
    return g
  
for l in fcooling:
    if not l.startswith("22-11-2013"):
    #if not l.startswith("04-12-2013"):
        continue
    print  l.strip().split()[0],l.strip().split()[1],l.strip().split()[2],l.strip().split()[3]
    #if l.strip().split()[0] is not '22-11-2013':
    #   continue
    #print l
    date = l.strip().split()[0]
    t = l.strip().split()[1]
    ti = l.strip().split()[3]
    hi = l.strip().split()[4]
    to = l.strip().split()[5]
    ho = l.strip().split()[6]
    Tdi = dewpoint_approximation(float(ti),float(hi))
    Tdo = dewpoint_approximation(float(to),float(ho))
    if Tdi < -40:Tdi=-40
    if Tdo < -40:Tdo=-40
    dewpointinside.append(Tdi)
    dewpointoutside.append(Tdo)
    tinside.append(float(ti))
    toutside.append(float(to))
    hinside.append(float(hi))
    houtside.append(float(ho))
    tt=datetime.strptime(t,"%H:%M:%S")
    difference=tt-starttime
    tdiff = difference.total_seconds()
    print tdiff
    times.append(tdiff/60)


    #print l.strip().split()[2]
```

```python
plt = figure(figsize=(10, 8))
plot(times,toutside,'--',label='T_outside')
plot(times,tinside,'--',label='T_inside')
plot(times,houtside,':',label='H_outside')
plot(times,hinside,':',label='H_inside')
plot(times,dewpointinside,label='DewPoint_inside')
plot(times
     ,dewpointoutside,label='DewPoint_outside')
#xlim(0,15)
xlabel('Time [min]',size = 15)
ylabel('Temperature(C)/Humidity',size = 15)
title('CO2 cooling system -- humidity test',size = 15)
legend(loc=0,bbox_to_anchor = (0.3, 0.7),fontsize=12)
savefig(outputlocation+'CO2test_1122_2014.pdf')
#xlim(15,50)
ylim(-40,50)
#fig, ax = subplots()
#line, = ax.plot([], [], lw=2)
#ax.set_ylim(-10, 60)
#ax.set_xlim(0,15)
#ax.set_
#line.set_data(times,houtside)
#line.set_data(times,hinside)

#subplots(111, xlim=(0.45,0.55), ylim=(0.4,.6),autoscale_on=False)
#plt.show()
```
